```python
# ดึงรหัสสถานี (Node) ที่ใช้ในโมเดลออกมา
# unique() ใช้เพื่อไม่ให้รหัสสถานีซ้ำกัน
stations_to_keep = Stations_in_the_model['Node'].unique()

# กรองข้อมูลฝนรายวันปี 2025
# เลือกเฉพาะแถวที่ tele_station_oldcode
# อยู่ในรายชื่อสถานีที่ใช้ในโมเดล
HII_daily_rainfall_2025_filtered = HII_daily_rainfall_2025[
    HII_daily_rainfall_2025['tele_station_oldcode'].isin(stations_to_keep)
].copy()

# แสดงขนาดข้อมูลก่อนกรอง (จำนวนแถว, จำนวนคอลัมน์)
print(f"Original HII_daily_rainfall_2025 shape: {HII_daily_rainfall_2025.shape}")

# แสดงขนาดข้อมูลหลังกรอง
print(f"Filtered HII_daily_rainfall_2025 shape: {HII_daily_rainfall_2025_filtered.shape}")

# แสดงข้อมูล 5 แถวแรกของ DataFrame หลังกรอง
HII_daily_rainfall_2025_filtered.head()
```

# ข้อที่ 3

```python
HII_daily_rainfall_2025_filtered.isnull().sum()
```

```python
# สร้างสำเนาของข้อมูลที่กรองแล้ว เพื่อไม่ให้กระทบต้นฉบับ
df = HII_daily_rainfall_2025_filtered.copy()

# แปลงคอลัมน์ measure_datetime ให้เป็นชนิด datetime
df['measure_datetime'] = pd.to_datetime(df['measure_datetime'])

# สร้างช่วงวันที่ต่อเนื่อง ตั้งแต่วันแรกถึงวันสุดท้ายของข้อมูล
dates = pd.date_range(
    df['measure_datetime'].min(),
    df['measure_datetime'].max()
)

# สร้างตารางที่มี "สถานี × ทุกวัน" ครบทุกความเป็นไปได้
# MultiIndex.from_product = Cartesian product
full = pd.MultiIndex.from_product(
    [df['tele_station_oldcode'].unique(), dates],
    names=['tele_station_oldcode', 'measure_datetime']
).to_frame(index=False)

# นำตาราง full มาต่อ (merge) กับข้อมูลจริง
# ถ้าวันไหนไม่มีข้อมูล rainfall_value จะได้ค่า NaN
missing = (
    full.merge(
        df[['tele_station_oldcode', 'measure_datetime', 'rainfall_value']],
        how='left'
    )
    # สร้างคอลัมน์ missing เป็น True/False
    .assign(missing=lambda x: x['rainfall_value'].isna())
    # นับจำนวนวันที่ข้อมูลหายไป แยกตามสถานี
    .groupby('tele_station_oldcode')['missing']
    .agg(missing_count='sum')
)

# คำนวณจำนวนวันที่ควรมีทั้งหมด และสัดส่วนข้อมูลที่หายไป
result = (
    missing
    .assign(
        total_expected=len(dates),
        missing_proportion=lambda x: x['missing_count'] / len(dates)
    )
    # เรียงจากสถานีที่ข้อมูลหายมากที่สุด
    .sort_values('missing_proportion', ascending=False)
)

# แสดงผลลัพธ์สุดท้าย
result
```

# ข้อที่ 4

```python
# ระบุรายชื่อสถานีที่ต้องตัดออก
# เงื่อนไข: สัดส่วนข้อมูลที่หายไป (missing_proportion) >= 20%
stations_to_remove = result[
    result['missing_proportion'] >= 0.20
].index.tolist()

# แสดงจำนวนและรายชื่อสถานีที่มีข้อมูลหายเกินเกณฑ์
print(f"Stations with >= 20% missing data records ({len(stations_to_remove)} stations):")
for station in stations_to_remove:
    print(f"- {station}")

# กรองข้อมูลฝนรายวัน
# เอาเฉพาะสถานีที่ "ไม่อยู่" ในรายชื่อที่ต้องตัดออก
HII_daily_rainfall_2025_cleaned = HII_daily_rainfall_2025_filtered[
    ~HII_daily_rainfall_2025_filtered['tele_station_oldcode'].isin(stations_to_remove)
].copy()

# นับจำนวนสถานีที่เหลืออยู่หลังการกรอง
remaining_stations_count = (
    HII_daily_rainfall_2025_cleaned['tele_station_oldcode']
    .nunique()
)

# แสดงสรุปผลการคัดกรอง
print(f"\nNumber of stations removed: {len(stations_to_remove)}")
print(f"Number of remaining stations after filtering: {remaining_stations_count}")
```

# ข้อที่ 5

```python
HII_daily_rainfall_2025_cleaned.isnull().sum()
```

```python
# สร้างสำเนาข้อมูลหลังผ่านการคัดสถานีแล้ว
df = HII_daily_rainfall_2025_cleaned.copy()

# แปลงคอลัมน์วันที่ให้เป็นชนิด datetime
df['measure_datetime'] = pd.to_datetime(df['measure_datetime'])

# สร้างตัวแปร day-of-year (1–365/366)
df['doy'] = df['measure_datetime'].dt.dayofyear


# คำนวณค่า median ของปริมาณฝน
# แยกตามสถานี และตามวันของปี (doy)
med = (
    df.groupby(['tele_station_oldcode', 'doy'])['rainfall_value']
      .median()
      .reset_index()
)


# สร้างช่วงวันที่ครบทุกวันตั้งแต่วันแรกถึงวันสุดท้าย
dates = pd.date_range(
    df['measure_datetime'].min(),
    df['measure_datetime'].max()
)

# สร้าง grid "สถานี × ทุกวัน"
full = pd.MultiIndex.from_product(
    [df['tele_station_oldcode'].unique(), dates],
    names=['tele_station_oldcode', 'measure_datetime']
).to_frame(index=False)

# เพิ่มตัวแปร day-of-year ให้กับ grid
full['doy'] = full['measure_datetime'].dt.dayofyear


# รวม grid กับข้อมูลจริง
# ถ้าวันไหนไม่มีข้อมูล rainfall_value จะเป็น NaN
df_imp = (
    full
    .merge(
        df[['tele_station_oldcode', 'measure_datetime', 'rainfall_value']],
        how='left'
    )
    # merge ค่า median รายสถานี–ราย doy
    .merge(
        med,
        on=['tele_station_oldcode', 'doy'],
        how='left',
        suffixes=('', '_median')
    )
)

# เติมค่า missing ของ rainfall_value
# ด้วยค่า median ของสถานีเดียวกันและวันเดียวกันของปี
df_imp['rainfall_value'] = (
    df_imp['rainfall_value']
    .fillna(df_imp['rainfall_value_median'])
)

# กำหนด DataFrame สุดท้ายหลัง imputation
HII_daily_rainfall_2025_median = df_imp
```

# ข้อที่ 6

ในการเตรียมข้อมูลปริมาณฝนก่อนนำไปวิเคราะห์หรือประยุกต์ใช้ในงาน Data Mining การจัดการกับข้อมูลที่ขาดหายถือเป็นขั้นตอนที่มีความสำคัญอย่างมาก โดยข้อมูลปริมาณฝนมีลักษณะเฉพาะที่แตกต่างจากข้อมูลเชิงปริมาณทั่วไป กล่าวคือ ในหลายช่วงเวลาปริมาณฝนมักมีค่าเป็นศูนย์หรือมีค่าต่ำ แต่ในบางวันอาจเกิดฝนตกหนักทำให้ค่าปริมาณฝนสูงมาก ส่งผลให้การกระจายของข้อมูลมีลักษณะแบบเบ้และมีค่าผิดปกติปรากฏอยู่เป็นจำนวนมาก ภายใต้ลักษณะข้อมูลเช่นนี้ การเลือกวิธีแทนค่าข้อมูลที่ขาดหายจึงต้องคำนึงถึงผลกระทบของค่าที่สูงผิดปกติเป็นหลัก

การใช้ค่ามัธยฐาน (Median) เป็นวิธีที่เหมาะสมในการแทนค่าข้อมูลที่ขาดหาย เนื่องจาก median เป็นค่าสถิติที่พิจารณาจากตำแหน่งกึ่งกลางของข้อมูล ไม่ได้รับอิทธิพลจากค่าที่สูงหรือต่ำผิดปกติเหมือนกับค่าเฉลี่ย (Mean) ดังนั้น แม้จะมีวันที่เกิดฝนตกหนักและให้ค่าปริมาณฝนสูงมาก ค่า median ก็ยังคงสะท้อนค่าปริมาณฝนปกติของข้อมูลส่วนใหญ่ได้ดีกว่า หากใช้ค่าเฉลี่ย ค่าที่ได้อาจสูงเกินความเป็นจริงและทำให้ค่าที่นำมาเติมไม่สอดคล้องกับพฤติกรรมจริงของปริมาณฝนในแต่ละวัน

นอกจากนี้ การใช้ค่า median ยังช่วยลดความบิดเบือนของโครงสร้างข้อมูลหลังการเติมค่า ทำให้การกระจายของข้อมูลไม่เปลี่ยนแปลงไปจากลักษณะเดิมมากนัก ซึ่งเป็นสิ่งสำคัญต่อการวิเคราะห์ข้อมูลในขั้นตอนถัดไป ไม่ว่าจะเป็นการวิเคราะห์เชิงสถิติ การหาความสัมพันธ์ของตัวแปร หรือการสร้างแบบจำลอง Data Mining หากข้อมูลถูกเติมด้วยค่าที่สูงหรือต่ำเกินจริง อาจส่งผลให้แบบจำลองเรียนรู้รูปแบบที่คลาดเคลื่อนและให้ผลลัพธ์ที่ไม่แม่นยำ

ดังนั้น การเลือกใช้ค่ามัธยฐาน (Median) ในการแทนค่าข้อมูลปริมาณฝนที่ขาดหายจึงเป็นวิธีที่เหมาะสม ช่วยให้ข้อมูลมีความสมเหตุสมผล ลดอิทธิพลของค่าผิดปกติ และคงลักษณะตามธรรมชาติของข้อมูลไว้ได้ ส่งผลให้ชุดข้อมูลมีคุณภาพสูงขึ้นและพร้อมสำหรับการนำไปวิเคราะห์หรือทำ Data Mining อย่างมีประสิทธิภาพ

```python
# แปลงคอลัมน์ measure_datetime ให้เป็นชนิด datetime
HII_daily_rainfall_2025_dropcol['measure_datetime'] = pd.to_datetime(
    HII_daily_rainfall_2025_dropcol['measure_datetime']
)

# สร้างตัวแปรปี (year) จากวันที่
HII_daily_rainfall_2025_dropcol['year'] = (
    HII_daily_rainfall_2025_dropcol['measure_datetime'].dt.year
)

# สร้างตัวแปรเดือน (month) จากวันที่
HII_daily_rainfall_2025_dropcol['month'] = (
    HII_daily_rainfall_2025_dropcol['measure_datetime'].dt.month
)

# รวมปริมาณฝนรายเดือน แยกตามสถานี ปี และเดือน
# ใช้ sum เพราะฝนรายเดือนคือผลรวมของฝนรายวัน
monthly_rainfall = (
    HII_daily_rainfall_2025_dropcol
    .groupby(['tele_station_oldcode', 'month', 'year'])['rainfall_value']
    .sum()
    .reset_index()
)

# เปลี่ยนชื่อคอลัมน์ให้สอดคล้องกับรูปแบบข้อมูลที่ต้องการ
monthly_rainfall.rename(columns={
    'tele_station_oldcode': 'Node',   # รหัสสถานี
    'rainfall_value': 'rain'          # ปริมาณฝนรายเดือน
}, inplace=True)

# เลือกเฉพาะคอลัมน์ที่ต้องการใช้งาน
output_monthly_rainfall = monthly_rainfall[['Node', 'rain', 'month', 'year']]

# แสดงข้อมูลตัวอย่าง 5 แถวแรก
output_monthly_rainfall.head()
```
